```python
import geopandas as gpd
import fiona
from matplotlib import pyplot as plt
```

### Get the data

```python
loans_filepath = "../../data/geospatial-learn-course-data/kiva_loans/kiva_loans/kiva_loans.shp"
world_loans = gpd.read_file(loans_filepath)
# world_loans.plot()
```

```python
# world = gpd.read_file("../../data/geospatial-learn-course-data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")
world = gpd.read_file("https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip")
```

### Plot the data

```python
world_plot = world.plot(figsize=(14,14), color='none', edgecolor='gainsboro', zorder=1)
world_loans.plot(ax=world_plot, markersize=1, color='darkorange', zorder=2)
```

### Select loans based in the Philippines

```python
PHL_loans = world_loans.loc[world_loans.country == "Philippines"].copy()
PHL_loans.plot()
```
